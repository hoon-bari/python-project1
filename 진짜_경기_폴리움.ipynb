```python
import requests
import json

# 행정구역 json파일
geo_path = '/content/gyeonggiBound.json'
kkd_si = json.load(open(geo_path, encoding = 'utf-8'))
kkd_si
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 한글 폰트 설정
plt.rcParams['font.family'] = 'NanumGothic'

plt.figure(figsize=(20, 12))
sns.barplot(x = df_gyeonggi['2022 반려동물 등록 수'], y = df_gyeonggi['지역구'])
plt.yticks(fontsize=12)
plt.xlabel("")
plt.ylabel("")
plt.show()
```

```python
geo_gu_list = [a['properties']['name'] for a in kkd_si['features']]
sorted(geo_gu_list), len(geo_gu_list)
```

```python
gyeonggi_data = df_gyeonggi.groupby('지역구', as_index = False)['구별 반려동물 등록비율'].sum()
gyeonggi_data
list(gyeonggi_data['지역구']), len(list(gyeonggi_data['지역구']))
```

```python
gyeongi_map = folium.Map(
    location=[37.559819, 126.963895],
    zoom_start=11,
    tiles='cartodbpositron'
)

folium.GeoJson(
    kkd_si,
    name='지역구'
).add_to(gyeongi_map)

gyeongi_map.choropleth(geo_data=kkd_si,
             data= gyeonggi_data,
             columns = gyeonggi_data.columns,
             fill_color='PuRd',
             fill_opacity=0.5,
             line_opacity=1,
             key_on = 'properties.name',
             legend_name="지역구별 인구 대비 반려동물 등록 비울"
            )
gyeongi_map.save('gyeongi_map2.html')
gyeongi_map
```
