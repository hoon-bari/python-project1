```python
import pandas as pd
df_seoul = pd.read_csv('/content/서울 파이널.csv')
df_seoul.drop('Unnamed: 0', axis = 1, inplace = True)
df_seoul.rename(columns = {'구별 반려동물 보유 비율' : '인구대비 반려동물 비율'}, inplace = True)
df_seoul
```

```python
# 폴리움 실행
import folium
```

```python
import requests
import json

# 서울 행정구역 json raw파일(githubcontent)
r = requests.get('https://raw.githubusercontent.com/southkorea/seoul-maps/master/kostat/2013/json/seoul_municipalities_geo_simple.json')
c = r.content
seoul_geo = json.loads(c)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 한글 폰트 설정
plt.rcParams['font.family'] = 'NanumGothic'
plt.figure(figsize=(20 , 12))
sns.barplot(x = df_seoul['2022 반려동물 등록 수'], y = df_seoul['지역구'])
# plt.rc('font', size=12)
# plt.title('서울시 반려동물 등록 수')
plt.xlabel("")
plt.ylabel("")
plt.show()
```

```python
import matplotlib.pyplot as plt 

plt.figure(figsize=(15, 15))
plt.rc('font', size=12)
plt.rc('xtick', labelsize=17)
colors = sns.color_palette('pastel')
plt.pie(df_seoul['인구대비 반려동물 비율'], labels = df_seoul['지역구'], autopct='%.1f%%')
# plt.title('서울 구별 인구대비 반려동물 등록 비율')
plt.show()
```

```python
seoul_group_data = df_seoul.groupby('지역구', as_index = False)['인구대비 반려동물 비율'].sum()
seoul_group_data
```

```python
geo_gu_list = [a['properties']['name'] for a in seoul_geo['features']]
sorted(geo_gu_list), len(geo_gu_list)
```

```python
seoul_group_data['지역구'] = seoul_group_data['지역구'].str.strip()
```

```python
list(seoul_group_data['지역구'])
```

```python
seoul_map = folium.Map(
    location=[37.559819, 126.963895],
    zoom_start=11,
    tiles='cartodbpositron'
)

folium.GeoJson(
    seoul_geo,
    name='지역구'
).add_to(seoul_map)

seoul_map.choropleth(geo_data=seoul_geo,
             data=seoul_group_data,
             columns = seoul_group_data.columns,
             fill_color='PuRd',
             fill_opacity=0.5,
             line_opacity=1,
             key_on='properties.name',
             legend_name="서울 구별 인구대비 반려동물 등록 비율"
            )
seoul_map.save('seoul_map2.html')
seoul_map
```
